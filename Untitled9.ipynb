```python
#  Image should be converted to be binary form i.e OTSU method

import numpy as np
import cv2 
from matplotlib import pyplot as plt
imag = cv2.imread('/content/grey.png',0)
blurred = imag
plt.imshow(blurred, 'gray')
plt.title('Input')
plt.show()
histogram = cv2.calcHist([blurred],[0],None,[256],[0,256])
plt.hist(blurred.ravel(),256,[0,256])
plt.show()
hist_normalize = histogram.ravel()/histogram.max()
Q = hist_normalize.cumsum() 
x_axis = np.arange(256) 
mini = np.inf 
thresh = -1
for i in range (1,256):
  p1,p2 = np.hsplit(hist_normalize,[i])
  q1,q2 = Q[i],Q[255]-Q[i]
  b1,b2 = np.hsplit(x_axis,[i])
  v1=255
  v2=255
  if (q1!=0 and q2!=0):
    m1,m2 = np.sum(p1*b1)/q1, np.sum(p2*b2)/q2
    v1,v2 = np.sum(((b1-m1)**2)*p1)/q1,np.sum(((b2-m2)**2)*p2)/q2
    fn = v1*q1 + v2*q2
    if fn < mini:
      mini = fn
      thresh = i


ret,binarized=cv2.threshold(blurred,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
isWritten = cv2.imwrite(r"C:\Users\Mihir\Desktop\project\written.jpg",binarized)
if isWritten:
  print('image is successfully saved as a file.') 
plt.imshow(binarized, 'gray')
plt.title('Output')
plt.show()
print ("Threshold gotten by the OpenCV implementation:",ret)
```
